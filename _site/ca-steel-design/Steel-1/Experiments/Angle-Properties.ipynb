## Compare tabulated area with that computed via 'flattened' angle

```python
import sst
SST = sst.SST()
```

```python
ls = SST.section_table('L')
ls.head()
```

```python
ls.columns
```

```python
lp = ls['A,D,B,T'.split(',')].copy()
lp.head()
```

```python
lp['FA'] = (lp['D'] + lp['B'] - lp['T'])*lp['T']
lp['PE'] = 100.*(lp['FA']/lp['A'] - 1.)
lps = lp.sort_values('PE',ascending=False)
lps
```
