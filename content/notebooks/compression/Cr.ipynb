```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
pi = np.pi
```

```python
Fy = 345.
E = 200000.
n = [0.93,1.34,2.24]
phi = 0.9
A = 1.
```

```python
def Fe(z):
    return pi*pi*E/z**2

def lamda(z):
    return np.sqrt(Fy/Fe(z))
```

```python
def Cr_S16(z,n=n):
    """S16-09 13.3.1"""
    return phi*A*Fy*(1 + lamda(z)**(2*n))**(-1./n)

def Cr_Euler(z):
    """Euler's elastic buckling."""
    return phi*A*Fy*lamda(z)**-2
```

```python
x = np.linspace(1,200)
x2 = np.linspace(60,200)
```

```python
plt.figure(figsize=(8,8))
ax = plt.axes()
ax.grid(True)
ax.set_xlim(0,250)
ax.title.set_text('Column Strength Comparison ($F_y={}$)'.format(int(Fy)))
for N in n:
    ax.plot(x,Cr_S16(x,n=N),label='S16-14: $n={}$'.format(N))
ax.plot(x2,Cr_Euler(x2),label='Euler elastic')
ax.xaxis.set_label_text('$KL/r$')
ax.yaxis.set_label_text('$F_r$, MPa')
ax.legend()
ax.text(130,200,r'$F_r=\frac{\phi F_y}{(1 + \lambda^{2n})^{1/n}}$',fontsize=20)
None
```
