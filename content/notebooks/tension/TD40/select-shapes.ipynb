## Alternative Tension Members for Example TD40

```python
# import useful utilities
from sst import SST
sst = SST()
from utils import show
```

### HSS Shapes

```python
hss = sst.section_tables(['HS'],'Mass')
hss.query('A>2500 & B==D').head(10)[['A','T','Mass']]
```

### Angle shapes (4 and 2 angles)

```python
hss = sst.section_tables(['L'],'Mass')
hss.query('A>2500/4 & Avl!="*"').head(10)[['A','T','Mass']]    # 4 angles
```

```python
hss.query('A>2500/2 & Avl!="*"').head(10)[['A','T','Mass']]     # 2 angles
```

### Channels

```python
hss = sst.section_tables(['C'],'Mass')
hss.query('A>2500/2 & Avl!="*"').head(10)[['A','Mass']]
```

### W Shapes

```python
hss = sst.section_tables(['W'],'Mass')
hss.query('A>2500 & Avl!="*"').head(10)[['A','Mass']]
```

### WT Shapes
These are not very practical, but possible if flange-bolted.

```python
hss = sst.section_tables(['WT'],'Mass')
hss.query('A>2500').head(10)[['Avl','A','T','Mass']]
```
